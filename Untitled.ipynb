```python
# Text cleaning function
def clean_text(text):
    if pd.isna(text): return ""
    text = str(text)
    text = re.sub(r'<[^>]+>', ' ', text)              # إزالة HTML
    text = re.sub(r'http\S+|www\.\S+', ' ', text)    # إزالة URLs
    text = re.sub(r'[\r\n]+', ' ', text)             # إزالة newlines
    text = re.sub(r"[^A-Za-z0-9\u0600-\u06FF\'\s]", " ", text)  # أحرف + أرقام + عربي + apostrophe
    text = text.lower()
    text = re.sub(r'\s{2,}', ' ', text).strip()
    return text

df['text_clean'] = df[text_col].apply(clean_text)

# إضافة عدد كلمات وعداد الحروف
df['token_count'] = df['text_clean'].str.split().str.len().fillna(0).astype(int)
df['char_count']  = df['text_clean'].str.len().fillna(0).astype(int)

df['avg_word_length'] = df['char_count'] / df['token_count'].replace(0,1)
df['punct_count'] = df['text_clean'].apply(lambda x: sum([1 for c in x if c in '.,;:!?']))
df['upper_count'] = df['text_clean'].apply(lambda x: sum([1 for w in x.split() if w.isupper()]))

# معاينة قبل وبعد
for i in range(5):
    print("Before:", df[text_col].iloc[i])
    print("After:", df['text_clean'].iloc[i])
    print("---")
```

```python


# Cell: Convert labels to numeric & remove missing safely


if label_col:
    # تحويل النصوص لـ 0 و 1
    label_mapping = {'negative': 0, 'positive': 1}
    df['label_numeric'] = df[label_col].map(label_mapping)

    # تحقق من القيم غير المحولة
    unmapped = df[df['label_numeric'].isna()][label_col].unique()
    if len(unmapped) > 0:
        print("Warning: Some labels could not be mapped:", unmapped)

    # إزالة النصوص الفارغة أو labels المفقودة
    df = df[df['text_clean'].notna() & df['text_clean'].str.strip().astype(bool)]
    df = df[df['label_numeric'].notna()]

    y = df['label_numeric'].values
    print("Number of missing labels after mapping:", pd.isna(y).sum())
else:
    print("Label column not found! Cannot create y.")
    y = None

```

```python
# نصوص فارغة
n_missing_text = df['text_clean'].eq('').sum()
# labels مفقودة
n_missing_label = df['label_numeric'].isna().sum() if 'label_numeric' in df.columns else None
# مكرر
n_duplicates = df.duplicated(subset=['text_clean']).sum()
# إحصائيات عدد الكلمات
token_stats = df['token_count'].describe()
# outliers (> mean + 3*std)
mean_tokens = df['token_count'].mean()
std_tokens  = df['token_count'].std()
high_cut = mean_tokens + 3*std_tokens
outliers_high = df[df['token_count'] > high_cut]

print("Total rows:", len(df))
print("Empty texts after cleaning:", n_missing_text)
print("Missing labels:", n_missing_label)
print("Duplicate texts:", n_duplicates)
print("Token count stats:")
print(token_stats)
print("Very long texts (> mean+3*std):", len(outliers_high))
print("Top 5 longest texts:")
print(df.nlargest(5, 'token_count')[['token_count','char_count','text_clean']])
```

```python
from sklearn.feature_extraction.text import CountVectorizer

# Top words
def top_n_words(texts, n=20):
    vectorizer = CountVectorizer(stop_words='english', max_features=20000)
    X = vectorizer.fit_transform(texts)
    sums = X.sum(axis=0)
    words = [(w, int(sums[0, idx])) for w, idx in vectorizer.vocabulary_.items()]
    words = sorted(words, key=lambda x: -x[1])[:n]
    return words

if 'label_numeric' in df.columns:
    for cls in df['label_numeric'].unique():
        texts = df[df['label_numeric']==cls]['text_clean'].dropna()
        print("Top words for class", cls)
        print(top_n_words(texts, n=20))
```

```python
#Logistic Regression Tuning
import optuna
from sklearn.model_selection import cross_val_score

def objective_lr(trial):
    C = trial.suggest_loguniform('C', 1e-3, 10)
    solver = trial.suggest_categorical('solver', ['liblinear', 'saga'])
    
    model = LogisticRegression(C=C, solver=solver, max_iter=1000, random_state=42)
    score = cross_val_score(model, X_train, y_train, cv=3, scoring='accuracy').mean()
    return score

study_lr = optuna.create_study(direction='maximize')
study_lr.optimize(objective_lr, n_trials=30)

print("Best params Logistic Regression:", study_lr.best_params)
print("Best CV Accuracy:", study_lr.best_value)
```

```python
#Random Forest Tuning
def objective_rf(trial):
    n_estimators = trial.suggest_int('n_estimators', 50, 500)
    max_depth = trial.suggest_int('max_depth', 5, 50)
    min_samples_split = trial.suggest_int('min_samples_split', 2, 10)
    min_samples_leaf = trial.suggest_int('min_samples_leaf', 1, 5)
    
    model = RandomForestClassifier(
        n_estimators=n_estimators,
        max_depth=max_depth,
        min_samples_split=min_samples_split,
        min_samples_leaf=min_samples_leaf,
        random_state=42,
        n_jobs=-1
    )
    score = cross_val_score(model, X_train, y_train, cv=3, scoring='accuracy').mean()
    return score

study_rf = optuna.create_study(direction='maximize')
study_rf.optimize(objective_rf, n_trials=30)

print("Best params Random Forest:", study_rf.best_params)
print("Best CV Accuracy:", study_rf.best_value)
```

```python
import joblib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.metrics import (accuracy_score, precision_score, recall_score, f1_score,
                             roc_auc_score, precision_recall_curve, auc,
                             confusion_matrix, ConfusionMatrixDisplay, brier_score_loss)
from sklearn.calibration import CalibrationDisplay

from xgboost import XGBClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression

# Keras
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping

# Best parameters
best_params_lr = { 'C': 2.47, 'solver': 'liblinear', 'max_iter': 1000, 'random_state':42 }   
best_params_rf = { 'n_estimators': 300, 'max_depth': None, 'min_samples_split':2,
                   'min_samples_leaf':1, 'max_features':'sqrt', 'random_state':42, 'n_jobs':-1 }
best_params_xgb = { 'n_estimators':160, 'max_depth':6, 'learning_rate':0.125,
                    'subsample':0.8, 'colsample_bytree':0.8, 'random_state':42, 'n_jobs':-1,
                    'use_label_encoder':False, 'eval_metric':'logloss' }

# DL architecture params 
best_params_dl = {
    'units1': 256,
    'units2': 128,
    'dropout1': 0.3,
    'dropout2': 0.2,
    'lr': 1e-3,
    'batch_size': 64,
    'epochs': 20
}

#  تحويل X إلى dense للـ DL وللاستخدام العام في بعض النماذج 
X_train_dense = X_train.toarray() if hasattr(X_train, "toarray") else X_train
X_test_dense  = X_test.toarray() if hasattr(X_test, "toarray") else X_test

# إعادة تدريب الموديلات بالبارامترات المختارة 
# 3.1 Logistic Regression
lr_final = LogisticRegression(**best_params_lr)
lr_final.fit(X_train, y_train)

# 3.2 Random Forest
rf_final = RandomForestClassifier(**best_params_rf)
rf_final.fit(X_train, y_train)

# 3.3 XGBoost
xgb_final = XGBClassifier(**best_params_xgb)
xgb_final.fit(X_train, y_train)

# 3.4 Dense NN (نبني وندرّب من الصفر مع نفس البارامترات)
def build_dl(params):
    model = Sequential([
        Dense(params['units1'], activation='relu', input_shape=(X_train_dense.shape[1],)),
        Dropout(params['dropout1']),
        Dense(params['units2'], activation='relu'),
        Dropout(params['dropout2']),
        Dense(1, activation='sigmoid')
    ])
    opt = tf.keras.optimizers.Adam(learning_rate=params['lr'])
    model.compile(optimizer=opt, loss='binary_crossentropy', metrics=['accuracy'])
    return model

dl_final = build_dl(best_params_dl)
early_stop = EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)

dl_final.fit(
    X_train_dense, y_train,
    validation_split=0.1,
    epochs=best_params_dl['epochs'],
    batch_size=best_params_dl['batch_size'],
    callbacks=[early_stop],
    verbose=2
)

# تقييم شامل على الـ test (احصائيات + ROC + PR + calibration)
models = {
    'Logistic Regression': lr_final,
    'Random Forest': rf_final,
    'XGBoost': xgb_final,
    'Dense NN': dl_final
}

results = []
plt.figure(figsize=(12,5))
# ROC subplot
plt.subplot(1,2,1)
for name, model in models.items():
    if name == 'Dense NN':
        probs = model.predict(X_test_dense).ravel()
    else:
        # بعض الموديلات قد لا تملك predict_proba (لكن هذه الثلاثة تملك)
        probs = model.predict_proba(X_test)[:,1]
    fpr = None
    try:
        from sklearn.metrics import roc_curve
        fpr, tpr, _ = roc_curve(y_test, probs)
        roc_auc = auc(fpr, tpr)
        plt.plot(fpr, tpr, label=f"{name} (AUC={roc_auc:.4f})")
    except Exception as e:
        print("ROC skipped for", name, e)
    # نحتفظ بالقيم لحساب PR و calibration بعد
    results.append((name, model, probs))

plt.plot([0,1],[0,1],'--', color='gray')
plt.title("ROC Curves")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.legend(loc='lower right')

# PR subplot
plt.subplot(1,2,2)
for name, model, probs in results:
    from sklearn.metrics import precision_recall_curve
    precision, recall, _ = precision_recall_curve(y_test, probs)
    pr_auc = auc(recall, precision)
    plt.plot(recall, precision, label=f"{name} (PR-AUC={pr_auc:.4f})")

plt.title("Precision-Recall Curves")
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.legend(loc='lower left')
plt.tight_layout()
plt.show()

# Detailed metrics table + confusion matrices + calibration
summary_rows = []
for name, model, probs in results:
    if name == 'Dense NN':
        y_pred = (probs > 0.5).astype(int)
    else:
        y_pred = model.predict(X_test)

    acc = accuracy_score(y_test, y_pred)
    prec = precision_score(y_test, y_pred)
    rec = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    try:
        roc_auc = roc_auc_score(y_test, probs)
    except:
        roc_auc = np.nan
    # PR AUC
    precision_vals, recall_vals, _ = precision_recall_curve(y_test, probs)
    pr_auc = auc(recall_vals, precision_vals)

    brier = brier_score_loss(y_test, probs)

    summary_rows.append({
        'Model': name,
        'Accuracy': acc,
        'Precision': prec,
        'Recall': rec,
        'F1': f1,
        'ROC_AUC': roc_auc,
        'PR_AUC': pr_auc,
        'Brier': brier
    })

    # Confusion matrix plot
    disp = ConfusionMatrixDisplay.from_predictions(y_test, y_pred, display_labels=['Negative','Positive'])
    disp.ax_.set_title(f"{name} Confusion Matrix")
    plt.show()

    # Calibration plot
    plt.figure(figsize=(6,4))
    CalibrationDisplay.from_predictions(y_test, probs, n_bins=10, strategy='uniform')
    plt.title(f"{name} Calibration Plot (Reliability Diagram)")
    plt.show()

# Comparison DataFrame
comparison_df = pd.DataFrame(summary_rows).set_index('Model')
print("\n=== Comparison Table ===")
display(comparison_df)

#  حفظ النماذج 
os.makedirs("models", exist_ok=True)
joblib.dump(lr_final, "models/logistic_final.joblib")
joblib.dump(rf_final, "models/rf_final.joblib")
joblib.dump(xgb_final, "models/xgb_final.joblib")
# save keras
dl_final.save("models/dl_final_keras.h5")

print("Models saved in ./models/")
```

```python
# خلية LIME جاهزة

import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from lime.lime_text import LimeTextExplainer
from scipy.sparse import hstack

# ----- نفس إعدادات التقسيم اللي استخدمتيها في التدريب الأصلي -----
TEST_SIZE = 0.2
RANDOM_STATE = 42
# -------------------------------------------------------------------

# 1) اقرأ الملف الخام
df = pd.read_csv("data\cleaned_reviews_day1.csv")
print("Columns in CSV:", df.columns.tolist())

# النصوص الخام (cleaned)
texts = df["text_clean"].astype(str).tolist()
labels = df["sentiment"].tolist()

# 2) أعدي نفس التقسيم
X_train_texts, X_test_texts, y_train, y_test = train_test_split(
    texts, labels, test_size=TEST_SIZE, random_state=RANDOM_STATE
    # , stratify=labels  # فكّي التعليق لو استعملتي stratify في الكود الأصلي
)

print(f"Number of test texts: {len(X_test_texts)}")
print("Example test text:", X_test_texts[0][:200])


# 4) دالة prediction متوافقة مع LIME
def predict_proba_for_lime(texts_list):
    # حوّل النصوص إلى sparse matrix
    vec = vectorizer.transform(texts_list)
    
    # أضف الأعمدة العددية (3 أعمدة ثابتة = صفر)
    extra_features = np.zeros((vec.shape[0], 3))  
    
    # دمج الاثنين
    final_input = hstack([vec, extra_features])
    
    return lr_model.predict_proba(final_input)

# 5) جهّزي LIME Explainer
class_names = [str(c) for c in getattr(lr_model, "classes_", [])] or ["0", "1"]
explainer = LimeTextExplainer(class_names=class_names)

# 6) اختاري نص من X_test_texts
idx = 3
sample_text = X_test_texts[idx]
print("\nUsing sample index:", idx)
print(sample_text)

# 7) شغلي LIME
exp = explainer.explain_instance(
    sample_text,
    predict_proba_for_lime,
    num_features=10
)

# 8) عرض النتائج
print("\nLIME explanation (as_list):")
print(exp.as_list())

# لو داخل نوتبوك
try:
    exp.show_in_notebook(text=True)
except:
    exp.save_to_file("lime_explanation.html")
    print("Saved explanation to lime_explanation.html")
```
